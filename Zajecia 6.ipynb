```python
#OPTYMALIZACJIA PLANU PRACOWNIKÓW

#Niestety nie udało mi się zautomatyzować procesu daletego niestety jest to zrobione ręcznie. 

# Import PuLP modeler function
import pulp
from pulp import*


prob = LpProblem("Plan", LpMaximize)

AniaPn=LpVariable("AniaPn",0,1)
AniaWt=LpVariable("AniaWt",0,1)
AniaSr=LpVariable("AniaSr",0,1)
AniaCzw=LpVariable("AniaCzw",0,1)
AniaPt=LpVariable("AniaPt",0,1)
StefanPn=LpVariable("StefanPn",0,1)
StefanWt=LpVariable("StefanWt",0,1)
StefanSr=LpVariable("StefanSr",0,1)
StefanCzw=LpVariable("StefanCzw",0,1)
StefanPt=LpVariable("StefanPt",0,1)
HektorPn=LpVariable("HektorPn",0,1)
HektorWt=LpVariable("HektorWt",0,1)
HektorSr=LpVariable("HektorSr",0,1)
HektorCzw=LpVariable("HektorCzw",0,1)
HektorPt=LpVariable("HektorPt",0,1)
OlafPn=LpVariable("OlafPn",0,1)
OlafWt=LpVariable("OlafWt",0,1)
OlafSr=LpVariable("OlafSr",0,1)
OlafCzw=LpVariable("OlafCzw",0,1)
OlafPt=LpVariable("OlafPt",0,1)
LidiaPn=LpVariable("LidiaPn",0,1)
LidiaWt=LpVariable("LidiaWt",0,1)
LidiaSr=LpVariable("LidiaSr",0,1)
LidiaCzw=LpVariable("LidiaCzw",0,1)
LidiaPt=LpVariable("LidiaPt",0,1)
PiotrPn=LpVariable("PiotrPn",0,1)
PiotrWt=LpVariable("PiotrWt",0,1)
PiotrSr=LpVariable("PiotrSr",0,1)
PiotrCzw=LpVariable("PiotrCzw",0,1)
PiotrPt=LpVariable("PiotrPt",0,1)

Ponsum=AniaPn + StefanPn + HektorPn + OlafPn + LidiaPn + PiotrPn 
Wtsum=AniaWt + StefanWt+ HektorWt + OlafWt + LidiaWt + PiotrWt 
Srsum=AniaSr + StefanSr+ HektorSr + OlafSr + LidiaSr + PiotrSr
Czwsum=AniaCzw + StefanCzw+ HektorCzw + OlafCzw + LidiaCzw + PiotrCzw
Ptsum=AniaPt + StefanPt + HektorPt + OlafPt + LidiaPt + PiotrPt 

Aniasum= AniaPn + AniaWt + AniaSr + AniaCzw + AniaPt
Stefansum= StefanPn +StefanWt + StefanSr + StefanCzw + StefanPt
Hektorsum= HektorPn +HektorWt + HektorSr + HektorCzw + HektorPt
Olafsum= OlafPn +OlafWt + OlafSr + OlafCzw + OlafPt
Lidiasum= LidiaPn +LidiaWt + LidiaSr + LidiaCzw + LidiaPt
Piotrsum= PiotrPn +PiotrWt + PiotrSr + PiotrCzw + PiotrPt




prob += Ponsum==2
prob += Wtsum==1
prob += Srsum==1
prob += Czwsum==1
prob += Ptsum==3

prob += Aniasum <=3
prob += Stefansum <=3
prob += Hektorsum <=3
prob += Olafsum <=3
prob += Lidiasum <=3
prob += Piotrsum <=3

prob += StefanWt + StefanSr + StefanCzw + StefanPt==0
prob +=  HektorCzw + HektorPt==0
prob += LidiaPn +LidiaWt + LidiaSr==0
prob +=  PiotrCzw + PiotrPt==0

prob += Aniasum * 100 + Stefansum * 50 + Hektorsum * 60 + Olafsum * 40 + Lidiasum * 11 + Piotrsum * 70 
                 
prob.solve()
print("Status:", LpStatus[prob.status])

for v in prob.variables():
    print(v.name, "=", v.varValue)
print("koszt ", value(prob.objective)) 
```
